```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Most Streamed Spotify Songs 2024.csv', encoding='unicode_escape')
```

```python
df.drop(['TIDAL Popularity'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df['Track Score']=df['Track Score'] .astype('int') 
```

```python
df['Track Score'].dtypes
```

```python
df['Spotify Popularity']=df['Spotify Popularity'].astype('int') 
```

```python
df['Apple Music Playlist Count'] = df['Apple Music Playlist Count'].astype('int') 
```

```python
df['Deezer Playlist Count'] = df['Deezer Playlist Count'].astype('int') 
```

```python
 df['Amazon Playlist Count'] = df['Amazon Playlist Count'] .astype('int') 
```

```python
df.columns
```

```python
df.rename(columns={'YouTube Playlist Reach':'YouTube Reach'})
```

```python
df.describe()
```

```python
df.columns
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'], errors='coerce')
# Create a new 'Year' column by extracting the year from the 'Date' column
df['Year'] = df['Release Date'].dt.year
```

```python
df.head()
```

```python
df.groupby(['Explicit Track'], as_index=False)['Track Score'].sum().sort_values(by='Track Score', ascending=False)
```

```python
ax =  sns.countplot(x = 'Explicit Track' , data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Track = df.groupby(['Explicit Track'], as_index=False)['Track Score'].sum().sort_values(by='Track Score', ascending=False)
sns.barplot(x = 'Explicit Track', y = 'Track Score' ,data=Track)
```

From the graph we can see that the Track score of songs are higher which are one that has curse words or language or art that is generally deemed sexual, violent, or offensive in nature.

```python
df.columns
```

```python
recent_10_years_df = df[df['Year'] >= (2024 - 10)]

# Create the count plot for the recent 10 years
plt.figure(figsize=(10, 6))
sns.countplot(data=recent_10_years_df, x='Year', hue='Explicit Track')
plt.title('All Time Rank of  Explicit Tracks by Year (Recent 10 Years)')
plt.xlabel('Year')
plt.ylabel('All Time Rank')
plt.xticks(rotation=45)
plt.legend(title='Explicit Track', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

From the visualization we can see that the year 2023  has the highest all time rank of songs in recent 10 years and in which the most somgs are songs which were not explicit .

```python
Artist_Track = df.groupby(['Artist'], as_index=False)['Track Score'].sum().sort_values(by='Track Score', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Artist', y = 'Track Score' ,data=Artist_Track)
plt.title('Top 10 Artist by their track Score ')
```

From the graph we can observe that the trackscore of Billie Eilish is higher among recent 10 Artist . 

```python
Artist_Explicit_Track = df.groupby(['Artist'], as_index=False)['Explicit Track'].sum().sort_values(by='Explicit Track', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Artist', y = 'Explicit Track' ,data=Artist_Explicit_Track)
plt.title('Top 10 Artist by their Explicit Score ')
```

```python
Album = df.groupby(['Album Name'], as_index=False)['Amazon Playlist Count'].sum().sort_values(by='Amazon Playlist Count', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Album Name', y = 'Amazon Playlist Count' ,data=Album)
plt.title('Top 10 Album and their  Amazon Playlist Count ')
```
